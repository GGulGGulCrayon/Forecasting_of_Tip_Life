```python
csv_file_list = glob.glob(os.path.join(os.getcwd(), "*.csv"))

result = pd.DataFrame()

for csv_list in csv_file_list:
    csv_file = pd.read_csv(csv_list)
    result = pd.concat([result, csv_file], axis = 0)
```

```python
result = result.reset_index().drop('index',axis = 1)
result['Event Time'] = pd.to_datetime(result['Event Time'])
Vibration = result.sort_values('Event Time')
```

```python
Vibration['Hour'] = Vibration['Event Time'].dt.hour
Vibration['Minute'] = Vibration['Event Time'].dt.minute
Vibration['Second'] = Vibration['Event Time'].dt.second
Electronic_current['Hour'] = Electronic_current['Event Time'].dt.hour
Electronic_current['Minute'] = Electronic_current['Event Time'].dt.minute
Electronic_current['Second'] = Electronic_current['Event Time'].dt.second
```

```python
Vibration.info()
```

```python
Vibration_mod1 = Vibration.groupby(by = ['Event Date', 'ShotNo', 'Hour', 'Minute', 'Second'])['Value'].mean()
Electronic_current_mod1 = Electronic_current.groupby(by = ['Event Date', 'ShotNo', 'Hour', 'Minute', 'Second'])['Value'].mean()
```

```python
Vibration_mod1.head(100)
```

```python
Vibration_mod1 = Vibration_mod1.reset_index()
Vibration_mod1['Date'] = Vibration_mod1['Date'] = Vibration_mod1['Event Date'] + ' ' + Vibration_mod1['Hour'].astype('str') + ':' + Vibration_mod1['Minute'].astype('str') + ':' +Vibration_mod1['Second'].astype('str')
Vibration_mod1['Date'] = pd.to_datetime(Vibration_mod1['Date'])
Electronic_current_mod1 = Electronic_current_mod1.reset_index()
Electronic_current_mod1['Date'] = Electronic_current_mod1['Event Date'] + ' ' + Electronic_current_mod1['Hour'].astype('str') + ':' + Electronic_current_mod1['Minute'].astype('str') + ':' + Electronic_current_mod1['Second'].astype('str')
Vibration_mod1.head()
```

```python
Electronic_current_mod1.info()
```

```python
Vibration_mod1.info()
```

```python
EC_V = pd.merge(Vibration_mod1, Electronic_current_mod1 , on = ['ShotNo', 'Hour', 'Minute', 'Second'], how = 'left')
EC_V
```

```python
EC_V = EC_V.rename({'Event Date_x' : 'Date', 'Value_x' : 'Vibration', 'Value_y' : 'EC', 'Date_x' : 'Total Date'}, axis = 1)
EC_V.drop(['Event Date_y', 'Date_y'], axis = 1, inplace = True)
```

```python
EC_V = EC_V.reset_index()
EC_V = EC_V.sort_values(by = ['Total Date', 'ShotNo', 'Hour', 'Minute', 'Second'], ascending = True)
EC_V = EC_V.set_index('Total Date')
```

```python
import random

shotno_list = list(EC_V['ShotNo'].unique())
shotno_list = sorted(shotno_list)
sample_shotno_list = shotno_list[0:10]

fig4 = plt.figure(figsize = (30,30))
for shotno in range(1, len(sample_shotno_list)+1):
    locals()['ax_{}'.format(shotno)] = fig4.add_subplot(len(sample_shotno_list),1, shotno)
    locals()['ax_assi_{}'.format(shotno)] = locals()['ax_{}'.format(shotno)].twinx()
    locals()['ax_{}'.format(shotno)].plot(EC_V[EC_V['ShotNo'] == shotno_list[shotno-1]]['EC'], color = 'b', label = 'EC')
    locals()['ax_{}'.format(shotno)].set_title('Shot_{}_EC_Vibration'.format(sample_shotno_list[shotno-1]), fontsize = 30)
    locals()['ax_{}'.format(shotno)].set_xticklabels(labels = EC_V['Date'],fontsize=20)
    locals()['ax_{}'.format(shotno)].set_yticklabels(labels = EC_V['EC'],fontsize=20)
    locals()['ax_assi_{}'.format(shotno)].plot(EC_V[EC_V['ShotNo'] == shotno_list[shotno-1]]['Vibration'], color = 'g', label = 'Vibration')
    locals()['ax_assi_{}'.format(shotno)].set_yticklabels(labels = EC_V['Vibration'],fontsize=20)
    locals()['ax_{}'.format(shotno)].legend(loc=0)
fig4.tight_layout()
fig4.show()
```
